# 종관기상관측

종관기상관측장비(ASOS)로부터 관측된 자료의 연(年)단위 파일을 조회하고 다운로드 받으실 수 있습니다.
종관기상관측이란 정해진 시각의 대기 상태를 파악하기 위해 모든 관측소에서 같은 시각에 실시하는 지상관측을 말하며, 관측방법은 기압, 기온, 풍향, 풍속, 상대습도, 강수량, 강수유무, 일사량, 일조시간, 지면온도, 초상온도, 지중온도, 토양수분, 지하수위 14개 요소에 대해서는 자동으로 관측하고, 시정, 구름, 증발량, 일기현상 등은 일부 자동과 목측(目測)으로 관측합니다.

## 2015.1.1~ 2017.12.31 의 시간당 날씨 데이터 조회

```python
df2015 = pd.read_csv('SURFACE_ASOS_112_HR_2015_2015_2016.csv', encoding="euc-kr")
df2016 = pd.read_csv('SURFACE_ASOS_112_HR_2016_2016_2017.csv', encoding="euc-kr")
df2017 = pd.read_csv('SURFACE_ASOS_112_HR_2017_2017_2018.csv', encoding="euc-kr")
df = pd.merge(df2015,df2016,how ='outer')
df = pd.merge(df,df2017,how='outer')
df
```

```python
df.columns
```

## 일조, 일사량에 따른 기온과 지면온도의 상관관계

```python
ex1_df = df.drop(df.columns[[0,4,5,6,7,8,9,10,13,14,15,16,17,18,19,20,21,23,24,25,26]],axis=1)
ex1_df
```

```python
ex1_df=ex1_df.fillna(0)
ex1_df
```

```python
ex1_df= ex1_df[(ex1_df['일조(hr)']>0)&(ex1_df['일사(MJ/m2)']>0)]
ex1_df
```

```python
ex_data = df
ex_data=ex_data.drop(ex_data.columns[[0,4,6,7,8,9]],axis =1)
ex_data
```

```python
ex_data.dtypes
```

```python
ex_data
i=0

new_data = pd.DataFrame(columns=['일시','기온(°C)','누적강수량(mm)','풍속(m/s)','일조(Sec)'])

temper = 0
wind = 0
rain = 0
sun = 0
time = 0

for row in ex_data:
    temper = temper + row['기온(°C)']
    wind = wind + n_data['풍속(m/s)']
    rain = n_data['누적강수량(mm)']
    sun = n_data['일조(Sec)']
    time = n_data['일시']
    i = i +1
    if i%1440==0 :
        temper = temper / 1440
        wind = wind /1440
        new_data.append([{'기온(°C)':temper,'풍속(m/s)': wind,'누적강수량(mm)':rain,'일조(Sec)':sun,'일시':time}])
        temper =0
        wind =0
        rain =0
        sun =0
        time =0


new_data
    
   
```
